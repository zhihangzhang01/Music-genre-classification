#### The validation performance for models

````python
import json
%matplotlib inline
from utils.plotting import plot_training

mlp_val_log = "logs//mlp//MLP_030524_204231.json"
cnn_val_log = "logs//cnn//CNN_030524_204600.json"
wav2vec2_log = "logs//wav2vec2//Wav2Vec2_030524_204800.json"
with open(mlp_val_log, 'r') as mlp:
    mlp_val_json = json.load(mlp)
with open(cnn_val_log, 'r') as cnn:
    cnn_val_json = json.load(cnn)
with open(wav2vec2_log, 'r') as wav2:
    wav2_val_json = json.load(wav2)
    
plot_training(mlp_val_json["train_losses"],
              mlp_val_json["train_accs"],
              mlp_val_json["val_losses"], 
              mlp_val_json["val_accs"],
              model_name="MLP",
              return_fig=False)

plot_training(cnn_val_json["train_losses"],
              cnn_val_json["train_accs"],
              cnn_val_json["val_losses"], 
              cnn_val_json["val_accs"],
              model_name="CNN",
              return_fig=False)

plot_training(wav2_val_json["train_losses"],
              wav2_val_json["train_accs"],
              wav2_val_json["val_losses"], 
              wav2_val_json["val_accs"],
              model_name="wav2vec2",
              return_fig=False)
````

````python
import json
import numpy as np
import pandas as pd

with open("logs//mlp//eval_logs//MLP_5.json") as f:
    mlp_log = json.load(f)

with open("logs//cnn//eval_logs//CNN_5.json") as f:
    cnn_log = json.load(f)
    
with open("logs//wav2vec2//eval_logs//Wav2Vec2_5.json") as f:
    wav2vec2_log = json.load(f)
    
results_df = pd.DataFrame({"model":["MLP", "CNN", "Wav2Vec2"], 
                    "accuracy":["{0:.2f}\u00B1 {1:.4f}".format(mlp_log["mean_acc"], np.sqrt(mlp_log["var_acc"])),
                                "{0:.2f}\u00B1 {1:.4f}".format(cnn_log["mean_acc"], np.sqrt(cnn_log["var_acc"])),
                                "{0:.2f}\u00B1 {1:.4f}".format(wav2vec2_log["mean_acc"], np.sqrt(wav2vec2_log["var_acc"]))]})

results_df
````
